```python
from selenium import webdriver

import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import json
import codecs
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
PAGE = 1
rating_page_url = "https://www.thegioididong.com/dtdd/iphone-11/danh-gia"
result = []
wd = webdriver.Firefox()

while True:
    url = rating_page_url + '?page=' + str(PAGE)
    print(f'Crawl data in page {PAGE}')
    wd.get(url)
    try:
        WebDriverWait(wd, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "par")))
    except TimeoutException:
        print("No more reviews")
    page = wd.page_source
    soup = BeautifulSoup(page, 'html.parser')
    check= soup.find("p", class_="cmt-top-name")
    comments = soup.find_all("li", class_="par")
    if check is not None:
        data = []
        for comment in comments:
            username = comment.find("p", class_="cmt-top-name").text
            star = comment.find('div', class_="cmt-intro").find('div', class_="cmt-top-star").find_all('i', class_='iconcmt-starbuy')
            content = comment.find("p", class_="cmt-txt").text
            data.append({"username": username, "content": content, "star": len(star)})
        result.append(data)
    else:
        break
    time.sleep(3)
    PAGE += 1  # Tăng giá trị của PAGE sau mỗi lần lặp

wd.close()
```

```python
# dump dict to json file
with codecs.open("data.json", "w", encoding='utf-8') as outfile:
    json.dump(result, outfile, ensure_ascii=False)
```

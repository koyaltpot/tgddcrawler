```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
age = [46,20,52,30,57,25,28,36,22,43,57,33,22,63,40,48,28,49]
bmi = [25.4,20.6,26.2,22.6,25.4,23.1,22.7,24.9,19.8,25.3,23.2,21.8,20.9,26.7,26.4,21.2,21.2,22.8]
chol = [3.5,1.9,4.0,2.6,4.5,3.0,2.9,3.8,2.1,3.8,4.1,3.0, 2.5,4.6,3.2,4.2,2.3,4.0]
d = {'age': age, 'bmi': bmi, 'chol': chol}
df = pd.DataFrame(data=d)
```

```python
sns.heatmap(df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
sns.pairplot(df, x_vars='age', y_vars='chol', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
X = df[['age']].values
y = df['chol'].values


# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
LR = LinearRegression().fit(X, y)
```

```python
input_age1 = 30
y_pred = LR.predict([[input_age1]])

print(f"Nồng độ chol ở tuổi {input_age1}: {y_pred[0]}")
```

```python
sns.pairplot(df, x_vars=['age', 'bmi'], y_vars='chol', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
X2 = df[['age', 'bmi']].values
y2 = df['chol'].values


# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
LR2 = LinearRegression().fit(X2, y2)
```

```python
input_age2 = 35
input_bmi2 = 25
y_pred = LR2.predict([[input_age2, input_bmi2]])

print(f"Nồng độ chol ở tuổi {input_age2} và BMI {input_bmi2}: {y_pred[0]}")
```
